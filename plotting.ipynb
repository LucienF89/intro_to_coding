```python
with open('student_data.txt', 'r') as f:
    data = f.readlines()

student_id = []
student_yr = []

# loop through each line in the dataset
for line in data[1:]:
    line = line.strip()     # remove the \ns

    # split each line into induividal data values
    split_line = line.split()

    # add the values to their specific lists (ids at index 0, yrs at index 1)
    student_id.append(split_line[0])
    student_yr.append(split_line[1])

# print(len(split_line))
# print(len(student_yr))

plt.scatter(student_id, student_yr)
plt.show()
```
